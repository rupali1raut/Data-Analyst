```python
a=df.groupby(["State","Year"])["MURDER"].max().reset_index()
a
```

```python
bars=sns.barplot(x="State",y="MURDER",data=a)
for bar in bars.patches:
    bars.bar_label(bars.containers[0])
bars.patches[0].set_facecolor('red')
plt.show()
```

```python
bars=sns.barplot(x="State",y="MURDER",hue="State",data=a)
for i, container in enumerate(bars.containers):
    bars.bar_label(container)
plt.show()
```
