Fetching data of MAX. of MURDER in each state.

```python
a=df.groupby(["State","Year"])["MURDER"].max().reset_index()
a
```

Creating a barplot fetching barplot of max. murder in each state along with city.



```python
bars=sns.barplot(x='State',y='MURDER',data=max_murder,color='red',hue="City")
for bar in bars.containers:
     bars.bar_label(bar)
plt.show()
```

** *Here we can conclude that Yavatmal city in MH state has highest murder cases(252) in 2002 comparing to others. while Serchhip city in mizoram is least highest.* **



---



Retriving Min murder from each state according to Murder ,City

```python
min_murder=df.groupby("State")[["MURDER","City"]].min()
min_murder
```

Creating a barplot fetching barplot of min. murder in each state along with city, as well as hue.

```python
bars=sns.barplot(x='State',y='MURDER',data=min_murder,hue="City",palette='pastel')
for bar in bars.containers:
     bars.bar_label(bar)
plt.show()
```

** *Here we can analyzed that ODISA state has NO murder cases (0) in 2002, while Garo hills east city has highest min. murder cases with count of 10.* **

Fetching Top 10 rape cases in MH State

```python
top_N_rape_mh=MH[["State","City","RAPE"]].sort_values(by="RAPE",ascending=False,ignore_index=True).head(10)
top_N_rape_mh
```

Creating barplot for Fetching Top 10 cities in rape cases in MH State with Pastel colour combination.

```python
bars=sns.barplot(x='City',y='RAPE',data=top_N_rape_mh,palette='pastel')
for bar in bars.containers:
     bars.bar_label(bar)
     plt.xticks(rotation=90)
plt.show()
```

**Here * we can analyzed that Mumbai city in MH state has topmost city in rape cases in 2002 comparing to others cities. * **

Finding State wise total IPC crimes in each state.

```python
total=df.groupby("State")["TOTAL IPC CRIMES"].sum().reset_index()
total
```

Creating barplot for total IPC crimes in each state.

```python
barp=sns.barplot(x="State",y="TOTAL IPC CRIMES", data=total,color="yellow")
for i in barp.containers:
  barp.bar_label(i)

plt.show()

```

Here we can analyzed that overall in india Rajasthan state has highest crime rate (including rape,murder,attempt to murder nd all) in 2002 comparing to others. while Nagaland state has lower crime rate.
